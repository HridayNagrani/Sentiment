```python
def Bert_Score(Sent):
    tokens = tokenizer.encode(Sent, return_tensors='pt')
    result = model(tokens)
    # A score between 1 and 5 is returned. 1 being the most negative and 5 being the most positive.
    return int(torch.argmax(result.logits))+1
```

```python
df['Bert_Sentiment'] = df['Content'].apply(lambda x: Bert_Score(x[:512]))
```

```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
analyser = SentimentIntensityAnalyzer()
```

```python
def Vader_Score(Sent):
    score = analyser.polarity_scores(Sent)
    # score between -1 and 1. -1 being the most negative and 1 being the most positive
    return score['compound']
```

```python
df['Vader_Sentiment'] = df['Content'].apply(lambda x: Vader_Score(x[:512]))
```

```python
df
```

```python
from textblob import TextBlob
def TextBlob_Score(Sent):
    # Score is between -1 to 1. -1 being the most negative and 1 being the most positive. 
    polarity = TextBlob(Sent).sentiment.polarity
    return polarity
```

```python
df['TextBlob_Sentiment'] = df['Content'].apply(lambda x: TextBlob_Score(x[:512]))
```
